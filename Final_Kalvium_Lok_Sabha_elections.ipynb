```python
import pandas as pd

# Read HTML tables from the URL
tables1 = pd.read_html('https://results.eci.gov.in/PcResultGenJune2024/index.htm')
data1 = tables1[0]
print(data1)
tables2 = pd.read_html('https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S01.htm')
data2 = tables2[0]
print('Andhrapradesh',data2)
tables3 = pd.read_html('https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S10.htm')
data3 = tables3[0]
print('Karnataka',data3)
tables4 = pd.read_html('https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S22.htm')
data4 = tables4[0]
print('Tamilnadu',data4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.pie(data1['Won'], labels=data1['Party'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Seats by Party')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(data2['Won'], labels=data2['Party'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Seats by Party in Andhrapradesh')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data3['Party'], data3['Won'], color='skyblue')
plt.xlabel('Party')
plt.ylabel('Won')
plt.title('Distribution of Seats by Party in Karnataka')
plt.xticks(rotation=45)
plt.tight_layout()  # Ensures labels fit nicely in the plot
plt.show()
```
